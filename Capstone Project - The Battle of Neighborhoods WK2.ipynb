### 2- We obtain List of postal codes of Canada

```python
#Show Postal_cod with their positions latitud and longitud
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
# Scrap List of postal codes of Canada wiki page content by using BeautifulSoup
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

<span style="font-family:Arial; font-size:10;">Convert content of PostCode HTML table as dataframe

```python
dfs = pd.read_html(tab)
toronto=dfs[0]
toronto.head()
# Change de columns Name
toronto.columns
toronto.columns= ['Postcode','Borough','Neighbourhood']
toronto.columns
```

<span style="font-family:Arial; font-size:10;">Clean dataframe</span>

```python
# clean dataframe 
# Ignore cells with a borough that is Not assigned

toronto = toronto[toronto.Borough!='Not assigned']
toronto = toronto[toronto.Borough!= 0]
toronto.reset_index(drop = True, inplace = True)
toronto.head()
```

<span style="font-family:Arial; font-size:10;">More than one neighborhood can exist in one postal code area</span>

```python
# More than one neighborhood can exist in one postal code area. Same Postcode that has more than one Neighborhoods, it joins in one row with all neighborhoods.  
i = 0
for i in range(0,toronto.shape[0]):
    if toronto.iloc[i][2] == 'Not assigned':
        toronto.iloc[i][2] = toronto.iloc[i][1]
        i = i+1
                                 
toronto_groupby = toronto.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
toronto_groupby.head()
toronto_groupby
```

<span style="font-family:Arial; font-size:10;">If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.</span>

```python
# Replacing the name of the neighbourhoods which are 'Not assigned' with value of Borough
toronto_groupby['Neighbourhood'].replace('Not assigned',toronto_groupby['Borough'],inplace=True)
toronto_groupby.shape
```

```python
radius=50000
LIMIT=50
VERSION = '20180605' # Foursquare API version
#url to fetch data from foursquare api
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&query=University'.format(
        CLIENT_ID, 
        CLIENT_SECRET,
        VERSION, 
        lat_Toronto, 
        lng_Toronto, 
        radius, 
        LIMIT)

# get all the data
results = requests.get(url).json()
venue_data=results["response"]['groups'][0]['items']
venue_details=[]
for row in venue_data:
    try:
        venue_id=row['venue']['id']
        venue_name=row['venue']['name']
        venue_category=row['venue']['categories'][0]['name']
        venue_lat=row['venue']['location']['lat']
        venue_lng=row['venue']['location']['lng']
        venue_details.append([venue_id,venue_name,venue_category,venue_lat,venue_lng])
    except KeyError:
        pass

column_names=['ID','Name','Category','Latitude','Longitude']
df_University_to = pd.DataFrame(venue_details,columns=column_names)
df_University_to.head()
```
